```python
import pandas as pd

file_path = 'bee-sbv-point.csv'
bee_data = pd.read_csv(file_path, encoding='cp949')

filtered_bee_data = bee_data[bee_data['축종'].isin(['벌-개량종', '벌-재래종'])]
filtered_bee_data
```

```python
filtered_bee_data = filtered_bee_data[['Latitude', 'Longitude', '축종', '발생두수']]
filtered_bee_data
```

```python
output_file_path = 'bee_filtered.csv'
filtered_bee_data.to_csv(output_file_path, index=False)
```
